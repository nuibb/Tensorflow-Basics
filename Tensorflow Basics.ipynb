# **Tensorflow Basics**

Closing session

```
import tensorflow as tf

node1 = tf.constant(5)
node2 = tf.constant(6)
node3 = tf.add(node1, node2)

sess = tf.Session()

print(sess.run([node1, node2]))
print(sess.run(node3))

sess.close()
```

Avoiding closing session

```
with tf.Session() as sess:
  output1 = sess.run([node1, node2])
  mulNode1 = node1 * node2
  output2 = sess.run(mulNode1)
  malNode2 = tf.math.multiply(node1, node2, name=None)
  output3 = sess.run(malNode2)
  
  print(output1)
  print(output2)
  print(output3)
```
